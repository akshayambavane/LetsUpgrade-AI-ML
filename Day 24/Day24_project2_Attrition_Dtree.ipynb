```python
import numpy as np
import pandas as pd
from sklearn import tree
from sklearn import preprocessing
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
```

#### Filling Null Values by mean and median

```python
df['TotalWorkingYears'] = df['TotalWorkingYears'].fillna(11.28) 
# 11.28 is the mean of TotalWorkingYears column 

df['NumCompaniesWorked'] = df['NumCompaniesWorked'].fillna(2.69)
# 2.69 is the median of NumCompaniesWorked
```

```python
df.isnull().sum()
```

### Dropping unnecessary columns

```python
df = df.drop(['EmployeeCount','EmployeeID', 'Over18','StandardHours'], axis=1)
df.head()
```

### Converting Text Column to Categorical Form

```python
from sklearn.preprocessing import LabelEncoder
def encode(cols):
    enc = LabelEncoder()
    for col in cols:
        df[col] = enc.fit_transform(df[col])
encode_cols = ['Attrition', 'BusinessTravel', 'Department', 'EducationField', 'Gender', 'JobRole', 'MaritalStatus']
encode(encode_cols)
df.head()
```

```python
df.columns
```

```python
X=df[['Age', 'BusinessTravel', 'Department', 'DistanceFromHome',
       'Gender', 'JobRole',
       'MaritalStatus', 'MonthlyIncome',
       'PercentSalaryHike', 'YearsSinceLastPromotion',
       'YearsWithCurrManager']]
X.head()
```

```python
Y=df['Attrition']
```

```python
Dt=DecisionTreeClassifier(max_depth=8)
```

```python
Dt.fit(X,Y)
```

```python
Dt.score(X,Y)
# score is depend on max_depth
```

### Finding important Variables with RandomForest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model=RandomForestClassifier(n_estimators=1000,max_features=2,oob_score=True)
```

```python
rf_model.fit(X,Y)
```

```python
print("OOB Accuracy:",rf_model.oob_score_)
```

```python
features=['Age', 'BusinessTravel', 'Department', 'DistanceFromHome',
       'Gender', 'JobRole',
       'MaritalStatus', 'MonthlyIncome',
       'PercentSalaryHike', 'YearsSinceLastPromotion',
       'YearsWithCurrManager']
```

```python
for feature,imp in zip(features,rf_model.feature_importances_):
    print(feature,imp)
```

### From above results Important variables are: Age, DistanceFromHome, MonthlyIncome, PercentageSalaryHike

```python
X=df[['Age', 'DistanceFromHome', 'MonthlyIncome', 'PercentSalaryHike']]
```

```python
Dt=DecisionTreeClassifier(max_depth=8)
```

```python
Dt.fit(X,Y)
```

```python
Dt.score(X,Y)
# score is depend on max_depth
```

### After finding important variable Accuracy is 88%
